### Simple Context Retrieval Function

```python
def retrieve_context(query, top_n=1):
    scored_docs = []
    for name, content in documents.items():
        score = query.lower() in content.lower()
        if score:
            scored_docs.append((name, content))

    # Return top matching document(s)
    if not scored_docs:
        return "No relevant context found."
    return "\n\n".join(doc for _, doc in scored_docs[:top_n])
```

```python
from sentence_transformers import SentenceTransformer
import os

markdown_folder = "/content/drive/MyDrive/Colab Notebooks/UWE Data"
raw_chunks = []

# Load and chunk files
for filename in os.listdir(markdown_folder):
    if filename.endswith(".md"):
        filepath = os.path.join(markdown_folder, filename)
        with open(filepath, "r", encoding="utf-8") as f:
            text = f.read()
            # Basic chunking by double newline or split every 200 words
            paragraphs = text.split("\n\n")
            for para in paragraphs:
                if len(para.strip()) > 20:
                    raw_chunks.append((filename, para.strip()))
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')  # Fast and good for semantic search

texts = [chunk for _, chunk in raw_chunks]
embeddings = model.encode(texts, show_progress_bar=True)
```

```python
import faiss
import numpy as np

dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))
```

```python
def semantic_retrieve_context(query, top_k=3):
    query_embedding = model.encode([query])
    distances, indices = index.search(np.array(query_embedding), top_k)

    results = []
    for idx in indices[0]:
        filename, chunk = raw_chunks[idx]
        results.append(f"From {filename}:\n{chunk}")

    return "\n\n".join(results)
```

```python
import gradio as gr
from transformers import pipeline
from sentence_transformers import SentenceTransformer, util
import markdown
import os
from glob import glob

# 1. Load Semantic Embedding Model
embedder = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
# 2. Load Markdown Files and Embed
markdown_dir = "markdown_folder"  # change this path to where your markdown files are
markdown_files = glob(os.path.join(markdown_dir, "*.md"))

documents = []
doc_texts = []

for file_path in markdown_files:
    with open(file_path, 'r', encoding='utf-8') as f:
        raw = f.read()
        html = markdown.markdown(raw)
        documents.append(raw)      # for display
        doc_texts.append(html)     # for embedding

corpus_embeddings = embedder.encode(doc_texts, convert_to_tensor=True)

# 3. Semantic Context Retriever
def retrieve_context(query, top_k=1):
    query_embedding = embedder.encode(query, convert_to_tensor=True)
    hits = util.semantic_search(query_embedding, corpus_embeddings, top_k=top_k)[0]
    return "\n\n".join([documents[hit["corpus_id"]] for hit in hits])

# 4. Load model
pipe = pipeline("text-generation", model="sentence-transformers/all-MiniLM-L6-v2", device_map="auto", trust_remote_code=True)
#pipe = pipeline("text-generation", model="tiiuae/falcon-7b-instruct", device_map="auto", trust_remote_code=True)
# 5. Chat history
chat_history = []

# 6. Chat interface logic
def chat_interface(query):
    context = retrieve_context(query)
    prompt = f"Context:\n{context}\n\nQuestion: {query}\nAnswer:"
    output = pipe(prompt, max_new_tokens=256, do_sample=False)[0]["generated_text"]
    answer = output.split("Answer:")[-1].strip()
    chat_history.append((query, answer))
    return chat_history

# 7. Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("## 🤖 UWE RAG Chatbot")
    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="Ask a question")
    clear = gr.Button("Clear Chat")

    def respond(message):
        chat = chat_interface(message)
        return "", chat

    msg.submit(respond, [msg], [msg, chatbot])
    clear.click(lambda: [], None, chatbot)

# 8. Launch the app
demo.launch()
```
